```python
# Solution 8
import numpy as np
import random
import math

# Monte Carlo Integration
def monte_carlo_integration(func, a, b, N):
    x = [random.uniform(a, b) for _ in range(N)]
    y = [func(x_i) for x_i in x]
    return sum(y) / N * (b - a)

# Define the function to integrate
def f(x):
    return x**3 + 3*x**2 + 3*x + 1

# Integrate from 50 to 100
a = 50
b = 100
N = 1000000
result = monte_carlo_integration(f, a, b, N)
print("Monte Carlo Integration Result:", result)

# Monte Carlo Simulation
def monte_carlo_simulation(probabilities, N):
    results = [random.choices(range(len(probabilities)), weights=probabilities, k=N)]
    return results

# Define the probabilities
probabilities = [0.2, 0.3, 0.5]

# Run the simulation
N = 10000
results = monte_carlo_simulation(probabilities, N)
print("Monte Carlo Simulation Results:", results)

# Markov Chain Monte Carlo
def markov_chain_monte_carlo(p, N):
    x = 0
    for _ in range(N):
        if random.random() < p:
            x = 1 - x
    return x

# Define the probability
p = 0.5

# Run the simulation
N = 10000
results = [markov_chain_monte_carlo(p, N) for _ in range(1000)]
print("Markov Chain Monte Carlo Results:", results)

# Monte Carlo Pi Estimation
def monte_carlo_pi(N):
    inside_circle = 0
    for _ in range(N):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        if math.sqrt(x**2 + y**2) <= 1:
            inside_circle += 1
    return 4 * inside_circle / N

# Run the simulation
N = 1000000
result = monte_carlo_pi(N)
print("Monte Carlo Pi Estimation Result:", result)
```

```python
# Solution 2 import numpy as np

import numpy as np

def characteristic_polynomial_and_eigenpairs(a):
    """
    Calculate the characteristic polynomial and eigenpairs for the matrix B.

    Parameters:
    a (numpy array): The matrix B = ((-1,-3), (4,6))

    Returns:
    p (numpy array): The characteristic polynomial
    eigenpairs (list of tuples): A list of tuples, where each tuple contains the eigenvalue and the corresponding eigenvector
    """
    # Convert the matrix to a NumPy array
    a = np.array(a)

    # Calculate the characteristic polynomial
    p = np.poly1d([1, -np.trace(a), np.linalg.det(a)])

    # Calculate the eigenpairs
    eigenpairs = np.linalg.eig(a)

    # Extract the eigenvalues and eigenvectors
    eigenvalues = eigenpairs[0]
    eigenvectors = eigenpairs[1]

    # Return the characteristic polynomial and eigenpairs
    return p, list(zip(eigenvalues, eigenvectors))

# Example usage:
a = np.array([[-1, -3], [4, 6]])
p, eigenpairs = characteristic_polynomial_and_eigenpairs(a)
print("Characteristic polynomial:", p)
print("Eigenpairs:", eigenpairs)
```
